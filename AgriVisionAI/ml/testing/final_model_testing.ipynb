```python
# ============================================================
# CELL 1
# CROP MODEL LOADING TEST
# ============================================================

import joblib
import pandas as pd


crop_model = joblib.load(
    "../models/crop_model.pkl"
)


print("Crop model loaded successfully")
print(type(crop_model))
```

```python
# ============================================================
# CELL 2
# CHECK CROP MODEL INPUT FEATURES
# ============================================================

crop_features = crop_model.feature_names_in_

print("Total Features:", len(crop_features))

for feature in crop_features:
    print(feature)
```

```python
# ============================================================
# CELL 3
# CREATE SAMPLE INPUT FOR CROP PREDICTION
# ============================================================

crop_test_input = pd.DataFrame(
    [[
        90,        # nitrogen
        42,        # phosphorus
        43,        # potassium
        20.87,     # temperature
        82.00,     # humidity
        6.50,      # ph
        202.93,    # rainfall

        1.068,     # nitrogen_scaled
        -0.344,    # phosphorus_scaled
        -0.101,    # potassium_scaled
        -0.935,    # temperature_scaled
        0.472,     # humidity_scaled
        0.043,     # ph_scaled
        1.810,     # rainfall_scaled

        175,       # npk_total
        0.514,     # nitrogen_ratio
        0.240,     # phosphorus_ratio
        0.245      # potassium_ratio
    ]],
    columns=crop_model.feature_names_in_
)


crop_test_input
```

```python
# ============================================================
# CELL 4
# TEST CROP MODEL PREDICTION
# ============================================================

crop_prediction = crop_model.predict(
    crop_test_input
)


print(
    "Predicted Crop Code:",
    crop_prediction[0]
)
```

```python
# ============================================================
# CELL 5
# CONVERT CROP CODE INTO CROP NAME
# ============================================================

import joblib


crop_encoder = joblib.load(
    "../models/crop_label_encoder.pkl"
)


crop_name = crop_encoder.inverse_transform(
    crop_prediction
)


print(
    "Recommended Crop:",
    crop_name[0]
)
```

```python
# ============================================================
# CELL 6
# FERTILIZER MODEL LOADING TEST
#
# Model:
# XGBoost Classifier
#
# Accuracy:
# 88.55 %
# ============================================================

import joblib
import pandas as pd


fertilizer_model = joblib.load(
    "../models/fertilizer_model.pkl"
)


print("Fertilizer model loaded successfully")
print(type(fertilizer_model))
```

```python
# ============================================================
# CELL 7
# CHECK FERTILIZER MODEL INPUT FEATURES
# ============================================================

fertilizer_features = (
    fertilizer_model.feature_names_in_
)


print(
    "Total Features:",
    len(fertilizer_features)
)


for feature in fertilizer_features:
    print(feature)
```

```python
# ============================================================
# CELL 8
# CREATE SAMPLE INPUT FOR FERTILIZER PREDICTION
# ============================================================

fertilizer_test_input = pd.DataFrame(
    [[0] * len(fertilizer_model.feature_names_in_)],
    columns=fertilizer_model.feature_names_in_
)


# Fill sample values

for column in fertilizer_test_input.columns:

    if "nitrogen" in column.lower():
        fertilizer_test_input[column] = 90

    elif "phosphorus" in column.lower():
        fertilizer_test_input[column] = 40

    elif "potassium" in column.lower():
        fertilizer_test_input[column] = 40

    elif "temperature" in column.lower():
        fertilizer_test_input[column] = 25

    elif "humidity" in column.lower():
        fertilizer_test_input[column] = 70

    elif "ph" in column.lower():
        fertilizer_test_input[column] = 6.5


fertilizer_test_input
```

```python
# ============================================================
# CELL 9
# TEST FERTILIZER MODEL PREDICTION
# ============================================================

fertilizer_prediction = fertilizer_model.predict(
    fertilizer_test_input
)


print(
    "Predicted Fertilizer Code:",
    fertilizer_prediction[0]
)
```

```python
# ============================================================
# CELL 10
# CONVERT FERTILIZER CODE INTO FERTILIZER NAME
#
# Purpose:
# Convert numerical model prediction
# into actual fertilizer recommendation.
# ============================================================

import joblib


fertilizer_encoder = joblib.load(
    "../models/fertilizer_label_encoder.pkl"
)


fertilizer_name = fertilizer_encoder.inverse_transform(
    fertilizer_prediction
)


print(
    "Recommended Fertilizer:",
    fertilizer_name[0]
)
```

```python
# ============================================================
# CELL 11
# PRICE MODEL LOADING TEST
#
# Model:
# Gradient Boosting Regressor
#
# Performance:
# R2   : 0.8413524895711193
# MAE  : 1006.9989593930792
# RMSE : 1880.1042785976144
# ============================================================

import joblib
import pandas as pd

price_model = joblib.load(
    "../models/price_model.pkl"
)

print("Price model loaded successfully")
print(type(price_model))
```

```python
# ============================================================
# CELL 12
# CHECK PRICE MODEL INPUT FEATURES
# ============================================================

price_features = price_model.feature_names_in_

print("Total Features:", len(price_features))

for feature in price_features:
    print(feature)
```

```python
# ============================================================
# CELL 13
# CREATE SAMPLE INPUT FOR PRICE PREDICTION
#
# Purpose:
# Generate test market data using the exact
# features required by Gradient Boosting model.
# ============================================================

price_test_input = pd.DataFrame(
    [[0] * len(price_model.feature_names_in_)],
    columns=price_model.feature_names_in_
)


# Fill realistic sample values

for column in price_test_input.columns:

    if column == "year":
        price_test_input[column] = 2026

    elif column == "month":
        price_test_input[column] = 7

    elif "state_code" in column:
        price_test_input[column] = 1

    elif "district_code" in column:
        price_test_input[column] = 10

    elif "market_code" in column:
        price_test_input[column] = 20

    elif "commodity_code" in column:
        price_test_input[column] = 5

    elif "variety_code" in column:
        price_test_input[column] = 2

    elif "grade_code" in column:
        price_test_input[column] = 1


price_test_input
```

```python
# ============================================================
# CELL 14
# TEST PRICE PREDICTION MODEL
# ============================================================

predicted_price = price_model.predict(
    price_test_input
)


print(
    "Predicted Market Price:",
    round(predicted_price[0], 2)
)
```
