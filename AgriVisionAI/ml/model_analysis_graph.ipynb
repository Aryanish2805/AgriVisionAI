```python
# ============================================================
# CELL 13
# CONFUSION MATRIX DATA PREPARATION
#
# Dataset  : Crop Recommendation
# Model    : Extra Trees Classifier
# Accuracy : 99.54545454545455 %
#
# Purpose:
# Evaluate the best crop prediction model
# by comparing actual crop classes with
# predicted crop classes.
#
# Target Variable:
# crop_code
# ============================================================


import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix



# ------------------------------------------------------------
# Load preprocessed crop dataset
# ------------------------------------------------------------

crop_df = pd.read_csv(

    "datasets/crop_recommendation_preprocessed.csv"

)



# ------------------------------------------------------------
# Separate input features and output label
# ------------------------------------------------------------

X_crop = crop_df.drop(

    columns=[

        "crop",
        "crop_code",
        "soil_condition"

    ]

)


y_crop = crop_df["crop_code"]



# ------------------------------------------------------------
# Recreate same testing split
# ------------------------------------------------------------

X_train_crop, X_test_crop, y_train_crop, y_test_crop = train_test_split(

    X_crop,

    y_crop,

    test_size=0.2,

    random_state=42,

    stratify=y_crop

)



# ------------------------------------------------------------
# Load Extra Trees trained model
# ------------------------------------------------------------

crop_model = joblib.load(

    "models/crop_extra_trees.pkl"

)



# ------------------------------------------------------------
# Generate predictions
# ------------------------------------------------------------

crop_prediction = crop_model.predict(

    X_test_crop

)



# ------------------------------------------------------------
# Create confusion matrix
# ------------------------------------------------------------

crop_cm = confusion_matrix(

    y_test_crop,

    crop_prediction

)



crop_cm
```

```python
# ============================================================
# CELL 14
#
# CONFUSION MATRIX VISUALIZATION
#
# Module   : Crop Recommendation System
# Dataset  : Crop Recommendation Dataset
# Model    : Extra Trees Classifier
# Accuracy : 99.54545454545455 %
#
# Objective:
# Analyze class-wise prediction performance and identify
# possible misclassification between crop categories.
#
# Interpretation:
# - Higher diagonal values indicate correct predictions.
# - Off-diagonal values represent classification errors.
#
# ============================================================


import matplotlib.pyplot as plt



# ------------------------------------------------------------
# Extract crop class names for better visualization
# ------------------------------------------------------------

crop_labels = (

    crop_df

    .sort_values("crop_code")

    ["crop"]

    .unique()

)



# ------------------------------------------------------------
# Create confusion matrix figure
# ------------------------------------------------------------

plt.figure(

    figsize=(14,12)

)



plt.imshow(

    crop_cm

)



plt.title(

    "Confusion Matrix Evaluation of Extra Trees Classifier\nfor Crop Recommendation System",

    fontsize=15,

    fontweight="bold"

)



plt.xlabel(

    "Predicted Crop Category",

    fontsize=12

)



plt.ylabel(

    "Actual Crop Category",

    fontsize=12

)



# ------------------------------------------------------------
# Add crop names on X and Y axis
# ------------------------------------------------------------

plt.xticks(

    range(len(crop_labels)),

    crop_labels,

    rotation=90,

    fontsize=8

)



plt.yticks(

    range(len(crop_labels)),

    crop_labels,

    fontsize=8

)



plt.colorbar(

    label="Number of Samples"

)



# ------------------------------------------------------------
# Display prediction counts inside matrix
# ------------------------------------------------------------

for i in range(crop_cm.shape[0]):


    for j in range(crop_cm.shape[1]):


        if crop_cm[i][j] > 0:


            plt.text(

                j,

                i,

                crop_cm[i][j],

                ha="center",

                va="center",

                fontsize=8

            )



plt.tight_layout()



# ------------------------------------------------------------
# Save high resolution figure for research report
# ------------------------------------------------------------

plt.savefig(

    "results/figures/07_crop_confusion_matrix_extra_trees.png",

    dpi=300,

    bbox_inches="tight"

)



plt.show()
```

```python
# ============================================================
# CELL 16
#
# ACTUAL VS PREDICTED DATA PREPARATION
#
# Module : Crop Price Prediction
# Model  : Gradient Boosting Regressor
#
# R²   : 0.8413524895711193
# MAE  : 1006.9989593930792
# RMSE : 1880.1042785976144
#
# Purpose:
# Use the exact training features saved inside
# the model to avoid feature mismatch.
#
# ============================================================


import joblib

from sklearn.model_selection import train_test_split



# ------------------------------------------------------------
# Load trained Gradient Boosting model
# ------------------------------------------------------------

price_model = joblib.load(

    "models/price_gradient_boosting.pkl"

)



# ------------------------------------------------------------
# Get exact features used during model training
# ------------------------------------------------------------

training_features = price_model.feature_names_in_


print(
    "Features used during training:"
)

print(
    training_features
)



# ------------------------------------------------------------
# Prepare dataset using same features
# ------------------------------------------------------------

X_price = price_df[

    training_features

]


y_price = price_df[

    "modal_price"

]



# ------------------------------------------------------------
# Recreate testing split
# ------------------------------------------------------------

X_train_price, X_test_price, y_train_price, y_test_price = train_test_split(

    X_price,

    y_price,

    test_size=0.2,

    random_state=42

)



# ------------------------------------------------------------
# Generate price predictions
# ------------------------------------------------------------

price_prediction = price_model.predict(

    X_test_price

)



print(

    "Price prediction completed successfully"

)
```

```python
# ============================================================
# CELL 17
#
# ACTUAL VS PREDICTED PRICE VISUALIZATION
#
# Module : Crop Price Prediction System
# Model  : Gradient Boosting Regressor
#
# Performance:
# R² Score : 0.8413524895711193
# MAE      : 1006.9989593930792
# RMSE     : 1880.1042785976144
#
# Objective:
# Compare actual market crop prices with
# machine learning predicted prices.
#
# Interpretation:
# Points closer to reference line indicate
# higher prediction accuracy.
#
# ============================================================



plt.figure(

    figsize=(9,9)

)



# ------------------------------------------------------------
# Plot actual prices vs predicted prices
# ------------------------------------------------------------

plt.scatter(

    y_test_price,

    price_prediction,

    alpha=0.5

)



# ------------------------------------------------------------
# Perfect prediction reference line
# ------------------------------------------------------------

minimum_price = min(

    y_test_price.min(),

    price_prediction.min()

)


maximum_price = max(

    y_test_price.max(),

    price_prediction.max()

)



plt.plot(

    [

        minimum_price,

        maximum_price

    ],


    [

        minimum_price,

        maximum_price

    ]

)



# ------------------------------------------------------------
# Graph labels and title
# ------------------------------------------------------------

plt.title(

    "Actual vs Predicted Crop Price Analysis\nUsing Gradient Boosting Regression Model",

    fontsize=14,

    fontweight="bold"

)



plt.xlabel(

    "Actual Crop Price",

    fontsize=12

)



plt.ylabel(

    "Predicted Crop Price",

    fontsize=12

)



plt.grid(

    alpha=0.3

)



plt.tight_layout()



# ------------------------------------------------------------
# Save research quality figure
# ------------------------------------------------------------

plt.savefig(

    "results/figures/08_price_actual_vs_predicted_gradient_boosting.png",

    dpi=300,

    bbox_inches="tight"

)



plt.show()
```

```python
# ============================================================
# CELL 19
#
# FERTILIZER FEATURE IMPORTANCE ANALYSIS
#
# Module : Fertilizer Recommendation System
# Model  : XGBoost Classifier
#
# Accuracy:
# 88.55 %
#
# Objective:
# Analyze important agricultural factors affecting
# fertilizer recommendation decisions.
#
# ============================================================


import pandas as pd
import matplotlib.pyplot as plt
import joblib
import os


# Load final fertilizer model

fertilizer_model = joblib.load(
    "models/fertilizer_model.pkl"
)


# Load dataset

fertilizer_df = pd.read_csv(
    "datasets/fertilizer_recommendation_preprocessed.csv"
)


# Use exact features used during training

fertilizer_features = (
    fertilizer_model
    .feature_names_in_
)


X_fertilizer = fertilizer_df[
    fertilizer_features
]


# Create feature importance dataframe

fertilizer_importance = pd.DataFrame(
    {
        "Feature": X_fertilizer.columns,
        "Importance": fertilizer_model.feature_importances_
    }
)


fertilizer_importance = (
    fertilizer_importance
    .sort_values(
        by="Importance",
        ascending=True
    )
)


# Plot

plt.figure(
    figsize=(10,8)
)


plt.barh(
    fertilizer_importance["Feature"],
    fertilizer_importance["Importance"]
)


plt.title(
    "Feature Importance Analysis\nXGBoost Classifier - Fertilizer Recommendation",
    fontsize=14,
    fontweight="bold"
)


plt.xlabel(
    "Importance Score",
    fontsize=12
)


plt.ylabel(
    "Input Features",
    fontsize=12
)


plt.grid(
    axis="x",
    alpha=0.3
)


plt.tight_layout()


# Save graph

os.makedirs(
    "results/figures",
    exist_ok=True
)


plt.savefig(
    "results/figures/10_fertilizer_feature_importance.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()


# Display feature ranking

fertilizer_importance.sort_values(
    by="Importance",
    ascending=False
)
```
