```python
import pandas as pd
import joblib

from sklearn.preprocessing import StandardScaler


df = pd.read_csv(
    "../datasets/crop_recommendation_preprocessed.csv"
)


scale_columns = [
    "nitrogen",
    "phosphorus",
    "potassium",
    "temperature",
    "humidity",
    "ph",
    "rainfall"
]


crop_scaler = StandardScaler()


crop_scaler.fit(
    df[scale_columns]
)


joblib.dump(
    crop_scaler,
    "../models/crop_scaler.pkl"
)


print(
    "crop_scaler.pkl saved successfully"
)
```
