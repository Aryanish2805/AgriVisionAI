```python
# ============================================================
# FINAL DEPLOYMENT MODEL PREPARATION
#
# Load already trained research models
# and save them with deployment names.
# ============================================================


import os
import joblib


os.makedirs(
    "models",
    exist_ok=True
)


print(
    os.listdir("models")
)
```

```python
# ============================================================
# Save Final Selected Deployment Models
# ============================================================

import joblib

# Crop Model
crop_model = joblib.load("models/crop_extra_trees.pkl")
joblib.dump(crop_model, "models/crop_model.pkl")


# Fertilizer Model
fertilizer_model = joblib.load(
    "models/fertilizer_xgboost.pkl"
)
joblib.dump(
    fertilizer_model,
    "models/fertilizer_model.pkl"
)


# Price Model
price_model = joblib.load(
    "models/price_gradient_boosting.pkl"
)
joblib.dump(
    price_model,
    "models/price_model.pkl"
)


print("Final deployment models saved successfully")
```

```python
import os

for file in [
    "crop_model.pkl",
    "fertilizer_model.pkl",
    "price_model.pkl"
]:
    print(
        file,
        "FOUND" if file in os.listdir("models") else "MISSING"
    )
```
